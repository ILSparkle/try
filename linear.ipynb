```python
#f = 2x^2 + 5x + 6
x = [0, 2, 5, -1, -2, -4]
y = [5.7, 23, 80, 2.7, 3, 17]
```

```python
a = 10
b = 1000
c = 2
m = 6
#f = ax2+bx+c
lr = 0.0001
```

```python
def cal_next(a0, b0, c0):
    sum_a = 0
    sum_b = 0
    sum_c = 0
    for i in range(0, m):
        sum_a = sum_a + (a0 * x[i] * x[i] + b0 * x[i] + c0 - y[i]) * x[i] * x[i]
        sum_b = sum_b + (a0 * x[i] * x[i] + b0 * x[i] + c0 - y[i]) * x[i]
        sum_c = sum_c + (a0 * x[i] * x[i] + b0 * x[i] + c0 - y[i])
    up_a = lr * sum_a / m
    up_b = lr * sum_b / m
    up_c = lr * sum_c / m
    return a0 - up_a, b0 - up_b, c0 - up_c
```

```python
def cal_loss(a0, b0, c0):
    loss = 0
    for i in range(0, m):
        loss = loss + (a0 * x[i] * x[i] + b0 * x[i] + c0 - y[i]) ** 2
    loss = loss / m
    return loss
```

```python
epochs = 300000
```

```python
for i in range(0, epochs):
    loss = cal_loss(a, b, c)
    if i % 3000 == 0:
        print("epoch - " + str(i) + "    loss:" + str(loss))
    a, b, c = cal_next(a, b, c)
    if i % 3000 == 0:
        print("a = " + str(a) + "  and b = " + str(b) + "  and c = " + str(c))
```
